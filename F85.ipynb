## Part I: Polarizers and Wave Plates

#### (1)
We measure the Brewster-Angly by polarizing the light in a circular way using a $\lambda$/4 wave plate. Then we use the the polarization filter (90°) and adjust the angle of the glass plate to measure the Brewster Angle. We get: $\theta_B = (40 \pm 5)^{\circ}$.

Reason why we can measure it like that: since the reflected light is always only poarized in one direction at the Brewster-Angle it gets blocked completely by using a polarization filter. That's why we can use this method to determine the angel when we see no light.

#### (2)
The light reflected after the first glass plate is linear polarized. From the Frenel equatinons we know that its direction is perpendicular to the plane of incidence. Since we put the plane of incindence of the second glass plate in this plane, nothing is reflected at the Brewster-Angle but everything transmitted.

#### (3)
Since the reflected beam is perpendicular polarized (to the plane of incidence) at the Brewster-Angle, we can use as a reference point to calibrate the second polarization filter.

The polarization axes for the two filters are at 96° and 99°.

#### (4)
The beam splitter (BS, not polarizing) isn't changing anything about the polarization of the two beams (as expected). The plarizing beam splitter (PBS) parts the beams similar like the glass plate at the Brewster-Angle. However, in this situation the transmitted beam is completely polarized while the reflected beam isn't completely polarized. So altogether we had two PBS and one BS.

#### (8)
The mirror acts like a $\lambda$/4 wave plate. With two perpendicular set-ups the results are the same.
It changes the linear polarized light to a eliptical polarized one.

#### (9)
First we reproduced the results from task 7 (case 1, Fig.3) und verified these.
After integrating the mirror the outgoing light was linear polarized in both cases. 
In the first case the plane of polarization got shifted for about 30° and in the second one for about 60°.

The mirror turns the circular polarized light, which hits it, to elliptical polarized light and gets turned into linear polarized light in the $\lambda$/4 wave plate. 
The polarization is then turned around compered to the initial polarization.

#### (10)
In the end the light is linear polarized, but shifted in 90°.

We are not certain though, if thats independent from the angle between the wave plate and the incident polarization.
A possible explanation could be, that if the light hits the wave plate in one of the two axis nothing happens, but the linear polarized light gets turned elliptical polarized one and when going trough the wave plate again, gets turned into linear polarized light again, which is shifted by 90° to the initial polarization.
If the light doesn't hit exactly one of the two axis,  then the circular polarized light doesn't get changed at all, which means a shift of the polarisation by 90°, like in Fig. 3 or 4.
